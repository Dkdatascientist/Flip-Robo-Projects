```python
from selenium import webdriver
import pandas as pd
import numpy as np
import time
```

```python
def get_url(search_term):
    """Generate a URL for search Term"""
    template = 'https://www.amazon.in/s?k={}&ref=nb_sb_noss_2'
    search_term = search_term.replace(' ','+')
    return template.format(search_term)

def search_top10_item(search_term):
    """Main Function"""
    #Start up web-driver.
    driver = webdriver.Chrome()
    url = get_url(search_term)
    
    driver.get(url)
    
    title = []
    Price = []
    Rating = []
    HREFF = []
    
    for i in driver.find_elements_by_xpath("//h2/a")[:10]:
        HREFF.append(i.get_attribute('href'))
    
    driver.close()
    
    for i in HREFF:
        driver = webdriver.Chrome()
        driver.get(i)
        time.sleep(2)
        
        title.append(driver.find_element_by_xpath("//h1[@id = 'title']").text)
        
        try:
            Price.append(driver.find_element_by_xpath("//span[@id = 'priceblock_dealprice']").text)
        except:
            try:
                Price.append(driver.find_element_by_xpath("//span[@id = 'priceblock_ourprice']").text)  
            except:
                Price.append("price not avaliable")
            
        driver.get(i+'#customerReviews')
        
        time.sleep(1)
        
        try:
            Rating.append(driver.find_element_by_xpath("//span[@class = 'a-size-medium a-color-base']").text)
            
        except:
            Rating.append(np.nan)
          
        driver.close()

    top_10_items = pd.DataFrame({"Title":title,"Ratings":Rating,"Price":Price})
    
    return top_10_items
```

```python
search_top10_item('Laptop Intel Core i7')
```

```python
search_top10_item('Laptop Intel Core i9')
```
